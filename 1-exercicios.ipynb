### 1. Ler e exibir um arquivo CSV usando pandas

```
import pandas as pd

df = spark.read.csv("/FileStore/tables/retail.csv", header=True, sep=';')
df = df.toPandas()
df.head()
```

### 2. Selecionar colunas específicas de um DataFrame usando pandas

```
# Filtrando a quantidade de produto vendido e identificador do cliente
df.filter(['quantity', 'customer_id'])
```

### 3. Filtrar linhas com base em uma condição usando pandas

```
# Filtrando regiões diferentes de UK e França
df.query('region != "United Kingdom" and region != "France"')
```

### 4. Agrupar dados com base em uma coluna usando pandas

```
# Data em que o pedido foi realizado agrupado com a quantidade de produtos vendido 
# Conseguimos visualizar o dia que teve a maior venda
df.groupby('invoice_date').agg(n=('quantity', 'count')).sort_values(by='n', ascending=False)
```

### 5. Usar a função max em alguma coluna do DataFrame

```
# Maior preço unitário e menor
df.agg({'unit_price': ['min', 'max']})
```
