### 1. Criar um array com 10 números inteiros aleatórios entre 1 e 50 e calcular a média e o desvio padrão

```
import numpy as np 

# Criar array com 10 números inteiros aleatórios entre 1 e 50, especificando a quantidade no 'size'
arr = np.random.randint(1, 51, size=10)

print('Array de números: ',arr)

# Calcular a média
media = np.mean(arr)

# Calcular o desvio padrão
desvio_padrao = np.std(arr)

print('\nMédia: ', media)
print(f'Desvio padrão: {desvio_padrao:.2f}')
```

### 2. Criar um array de 10 números aleatórios e calcular a raiz quadrada de todos os valores menores ou iguais a 0,5

```
# Criar array com 10 números aleatórios
arr = np.random.rand(10)

print('Array original: ', arr)

# Filtrar os números menores ou igual 0.5
numeros_filtrados = arr[arr <= 0.5]

# Calcular a raiz quadrada
raizes_quadradas = np.sqrt(numeros_filtrados)

print('\nNúmeros filtrados:', numeros_filtrados)
print('\nRaízes quadradas dos números filtrados:', raizes_quadradas)
```

### 3. Criar uma matriz 3D de tamanho 3x3x3 com valores aleatórios e exibir o valor máximo de cada linha de cada matriz 2D dentro dessa matriz 3D

```
# Criar matriz 3D 3x3x3x com valores aleatórios 
matriz_3d = np.random.rand(3, 3, 3)

print('Matriz 3D:')
print(matriz_3d)

# Calcular o valor máximo de cada coluna de cada matriz 2D
valores_maximos = np.max(matriz_3d, axis=1)

print('\nValor máximo de cada linha de cada matriz 2D:')
print(valores_maximos)
```

### 4. Criar uma matriz de tamanho 5x5 com valores aleatórios e subtrair a média de cada linha dessa matriz

```
# Criar matriz 
matriz = np.random.rand(5, 5)

print('Matriz original:')
print(matriz)

# Calcular a média
# axis = 1 -> Linha
# axis = 0 -> Coluna
medias = np.mean(matriz, axis=1)

# Subtrarir a média de cada linha
matriz_media_subtraida = matriz - medias[:, np.newaxis]

print('\nMédia de cada linha: ', medias)
print('\nMatriz com a média subtraída:')
print(matriz_media_subtraida)
```

### 5. Criar uma matriz de tamanho 4x4 com valores aleatórios e calcular a média e o desvio padrão das diagonais da matriz

```
# Criar matriz 
matriz = np.random.rand(4,4)

print('Matriz:') 
print(matriz)

diag_principal = np.diag(matriz)
diag_secundaria = np.diag(np.fliplr(matriz))

# Calcular a média
media_diagonal = np.mean([diag_principal, diag_secundaria])

# Calcular o desvio padrão
desvio_diagonal = np.std([diag_principal, diag_secundaria])

print('\nDiagonal principal: ', diag_principal)
print('\nDiagonal secundária: ', diag_secundaria)
print('\nMédia das diagonais: ', media_diagonal)
print('\nDesvio padrão das diagonais: ', desvio_diagonal)
```
