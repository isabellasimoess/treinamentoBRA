### 1. Some todos os números de 1 até 100

```python
soma = 0 

for numero in range(1,101):
  soma += numero
  
  print(f'A soma de todos os números de 1 até {numero} é:', soma)
```

### 2. Some todos os números pares de 1 até 100

```python
soma = 0 

for numero in range(1,101):
  if numero % 2 == 0:  # Verifica se o número é par
    soma += numero
    
    print(f'A soma de todos os números de 1 até {numero} é:', soma)
```

### 3. Some todos os números ímpares de 1 até 100

```python
soma = 0 

for numero in range(1,101):
  if numero % 2 != 0: # Verifica se o número é ímpar
    soma += numero
    
    print(f'A soma de todos os números de 1 até {numero} é:', soma)
```

### 4. Dada a lista abaixo, some todos os elementos inteiros positivos

```python
lista = [-2, 2, 'celular', 10, -99, False]
lista
```

```python
soma = 0

for elemento in lista:
  if (isinstance(elemento, int) == True ) and (elemento > 0):
    soma += elemento
    
print('A soma dos elementos é:', soma)
```
